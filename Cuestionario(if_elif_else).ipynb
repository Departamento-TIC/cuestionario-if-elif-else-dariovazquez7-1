````python
print ("¡A jugar!")

contador = 0

respuesta = int(input("Cuanto es 5*(3+2)="))

if respuesta == 25:
  print("Correcto")
  contador += 1
else:
  print("Incorrecto")

respuesta = int(input("Cuanto es 5*4+1="))

if respuesta == 21:
  print("Correcto")
  contador +=1
else:
  print("Incorrecto")

respuesta = input("Cual es el planeta mas cercano al sol:")

if respuesta == "mercurio":
  print("Correcto")
  contador +=1
else:
  print("Incorrecto")

print("¿Que pais no pertenece a Europa?: ")

print("1. España")
print("2. Italia")
print("3. Marruecos")
print("4. Francia")
respuesta = input("? ")

if respuesta == "3":
  print("Correcto")
  contador += 1
else:
  print("Incorrecto")


print("¿Quién no juega al fútbol?: ")

print("1. Messi")
print("2. Lebron James")
print("3. Ronaldo")
print("4. Benzema")
respuesta = input("? ")

if respuesta == "2":
  print("Correcto")
  contador += 1
else:
  print("Incorrecto")


print(f"Mis felicitaciones, has conseguido {contador} respuestas correctas!")
porcentaje = (contador /5) * 100
print(f"Eso es una puntuación del {porcentaje} por ciento.")
````
